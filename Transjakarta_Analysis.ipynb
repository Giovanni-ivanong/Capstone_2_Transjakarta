Sekarang kita mulai proses Data Cleaning.
Langkah pertama yang akan kita lakukan adalah mengisi missing values pada kolom `corridorID` dan `corridorName`.

Karena kedua kolom ini saling berkaitan—dimana setiap `corridorID` memiliki pasangan `corridorName` yang unik—maka kita bisa memanfaatkan data yang sudah ada untuk mengisi nilai yang hilang.

- Jika `corridorName` kosong, kita isi berdasarkan `corridorID` yang sesuai.
- Sebaliknya, jika `corridorID` kosong, kita isi berdasarkan nilai `corridorName`.

Dengan cara ini, kita bisa meminimalkan kehilangan data dan tetap menjaga konsistensi informasi.

```python
# Pertama, kita akan membuat pemetaan dari data yang tersedia untuk CorridorID dan CorridorName.
corridor_id_to_name = original_data.dropna(subset=['corridorID', 'corridorName']).drop_duplicates('corridorID').set_index('corridorID')['corridorName'].to_dict()
corridor_name_to_id = original_data.dropna(subset=['corridorID', 'corridorName']).drop_duplicates('corridorName').set_index('corridorName')['corridorID'].to_dict()
```

```python
# Memperbaiki case sensitive huruf nama kolom dan mencoba mengisi lagi
original_data['corridorName'] = original_data.apply(
    lambda row: corridor_id_to_name.get(row['corridorID'], row['corridorName']) if pd.isna(row['corridorName']) else row['corridorName'],
    axis=1
)

# Verifikasi koreksi dan periksa nilai yang hilang untuk kolom-kolom ini setelah diisi
original_data[['corridorID', 'corridorName']].isna().sum()
```

Selanjutnya, kita akan menangani kolom-kolom yang saling terkait dalam proses tap out, yaitu:

- `tapOutStops`

- `tapOutStopsName`

- `tapOutStopsLat`

- `tapOutStopsLon`

- `stopEndSeq`

- `tapOutTime`

Nilai yang hilang pada kolom-kolom ini umumnya terjadi bersamaan dalam satu baris, sehingga pendekatan yang digunakan adalah mencari baris lain dengan data lengkap yang memiliki kemiripan (misalnya berdasarkan `payCardID`, `tapInTime`, atau `tapInStops`) dan menggunakannya sebagai referensi untuk mengisi kekosongan tersebut.

Langkah ini bertujuan untuk menjaga konsistensi dan integritas data sebelum dilakukan analisis lebih lanjut.

```python
# Group related columns for tapOut information
tapout_columns = ['tapOutStops', 'tapOutStopsName', 'tapOutStopsLat', 'tapOutStopsLon', 'stopEndSeq', 'tapOutTime']

# Create a complete case subframe to use for imputation
complete_tapout = original_data.dropna(subset=tapout_columns)

# Define a function to fill missing values based on available complete cases
def fill_missing_tapout(row, reference_df, columns):
    if pd.isna(row['tapOutStops']):
        similar_rows = reference_df[reference_df['tapInStops'] == row['tapInStops']]
        if not similar_rows.empty:
            for col in columns:
                row[col] = similar_rows[col].iloc[0]
    return row

# Apply the function to fill missing values
original_data = original_data.apply(fill_missing_tapout, axis=1, args=(complete_tapout, tapout_columns))

# Check the remaining missing values for these tapOut columns after filling
original_data[tapout_columns].isna().sum()
```

Setelah mencoba mengisi berdasarkan data yang mirip namun masih terdapat nilai hilang, maka langkah selanjutnya adalah menggunakan pendekatan statistik sederhana:

- Median digunakan untuk kolom numerik: `tapOutStopsLat`, `tapOutStopsLon`, `stopEndSeq` karena median lebih tahan terhadap outlier.
- Modus digunakan untuk kolom kategorikal: `tapOutStops`, `tapOutStopsName`, `tapOutTime` karena mewakili nilai yang paling umum muncul.

Metode ini merupakan solusi standar ketika informasi yang lebih spesifik tidak tersedia, dan bertujuan untuk menjaga kelengkapan data tanpa mengubah distribusi secara signifikan.

```python
# For numerical columns, fill with median
numerical_columns = ['tapOutStopsLat', 'tapOutStopsLon', 'stopEndSeq']
for col in numerical_columns:
    median_value = original_data[col].median()
    original_data[col].fillna(median_value, inplace=True)

# For categorical columns, fill with mode
categorical_columns = ['tapOutStops', 'tapOutStopsName', 'tapOutTime']
for col in categorical_columns:
    mode_value = original_data[col].mode()[0]
    original_data[col].fillna(mode_value, inplace=True)

# Check the remaining missing values for these tapOut columns after filling
original_data[tapout_columns].isna().sum()
```

Kita akan mulai dengan `tapInStops` terlebih dahulu. Karena `tapInStops` berkaitan erat dengan lokasi dan identitas halte, kita dapat memanfaatkan informasi dari kolom:

`tapInStopsName`

`tapInStopsLat`

`tapInStopsLon`

Dengan mencocokkan koordinat lokasi dan nama halte, kita dapat mengisi nilai `tapInStops` berdasarkan halte yang paling mendekati secara spasial dan identitas.
Metode ini memungkinkan pengisian nilai hilang secara lebih akurat dan masuk akal, dibandingkan menggunakan metode acak atau umum seperti modus.

```python
# Creating a mapping from the combination of 'tapInStopsName', 'tapInStopsLat', and 'tapInStopsLon' to 'tapInStops'
tap_in_stops_mapping = original_data.dropna(subset=['tapInStops']).drop_duplicates(
    subset=['tapInStopsName', 'tapInStopsLat', 'tapInStopsLon']).set_index(
    ['tapInStopsName', 'tapInStopsLat', 'tapInStopsLon'])['tapInStops'].to_dict()

# Function to fill missing 'tapInStops' based on the mapping
def fill_tap_in_stops(row):
    if pd.isna(row['tapInStops']):
        return tap_in_stops_mapping.get((row['tapInStopsName'], row['tapInStopsLat'], row['tapInStopsLon']))
    return row['tapInStops']

# Apply the function to fill missing 'tapInStops'
original_data['tapInStops'] = original_data.apply(fill_tap_in_stops, axis=1)

# Check the remaining missing values for 'tapInStops' after filling
original_data['tapInStops'].isna().sum()
```

Untuk mengisi 34 nilai yang hilang pada kolom tapInStops, pendekatan yang digunakan adalah dengan memperkirakan halte masuk berdasarkan kedekatan geografis terhadap tapOutStops.



```python
from scipy.spatial.distance import cdist

# Extract known stops with coordinates
known_stops = original_data.dropna(subset=['tapInStops', 'tapInStopsLat', 'tapInStopsLon'])
coordinates = known_stops[['tapInStops', 'tapInStopsLat', 'tapInStopsLon']].drop_duplicates()

# For missing 'tapInStops', calculate the closest stop based on 'tapOutStopsLat' and 'tapOutStopsLon'
def find_closest_stop(row):
    if pd.isna(row['tapInStops']):
        # Calculate distances to all known stops
        distances = cdist([(row['tapOutStopsLat'], row['tapOutStopsLon'])],
                          coordinates[['tapInStopsLat', 'tapInStopsLon']],
                          metric='euclidean')
        # Find the index of the minimum distance
        min_index = distances.argmin()
        return coordinates.iloc[min_index]['tapInStops']
    return row['tapInStops']

# Apply the function to fill missing 'tapInStops'
original_data['tapInStops'] = original_data.apply(find_closest_stop, axis=1)

# Verify the filling
original_data['tapInStops']
```

```python
# Fill missing values in 'PayAmount' with the median
median_pay_amount = original_data['payAmount'].median()
original_data['payAmount'].fillna(median_pay_amount, inplace=True)

# Verify changes
original_data.info()
```
